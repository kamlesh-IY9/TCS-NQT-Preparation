''' 1. ArrAY 
    q.1 Given an array of integers, move all 0’s to the end while maintaining the order 
        of non-zero elements.
        
        Example:

            Input:  [4, 0, 5, 0, 3, 0, 1]

            Output: [4, 5, 3, 1, 0, 0, 0]
'''


```python
# a = [4, 0, 5, 0, 3, 0, 1]
a = [0,0,0,1]
# [1,2,3]
# [0,0,0]
ii = 0
for i in range(len(a)) :
    print("i = ",i)
    if a[i] != 0:
        a[i], a[ii] = a[ii], a[i]
        print(a[i], a[ii])
        ii += 1
print(a)
```
